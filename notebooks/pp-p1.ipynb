### Project 1: The TO-DO List 

---

*THE DNA OF BACKEND ENGINEERING*


- `Goal`: Build a program where users can add
tasks (e.g.,"Finish Python assignment")
to a list and view them.
- `Key Skills`: Control flow, decision-making logic, basic AI concepts.

```python
task_count = int(input('How many tasks do you want to list? \n Enter here : '))

my_tasks = []
for i in range(task_count):
    task = input(f'Enter task {i+1} : ')
    my_tasks.append(task)

print('Your TO-DO List: ')
for idx, task in enumerate(my_tasks):
    print(idx, task)
```
