### Project 2: Expense Tracker 

---

- `Goal`: Create a script where user enter an expense amount (e.g., 100, 50, 20). The program adds them up and dispalys the total spent.
- `Key Skills`: Math Operations & Accumulators (total = total + new_expense)
- `Why it matters`: Essential for understanding data processing and calculations in backend logic.

```python
curr_amount = 1000
print(f"Initial balance: $ {curr_amount}")
while True:
    try:
        user_input = input("Enter expense amount (type q to quit): ")
        if user_input.lower() == "q":
            break
        expense = int(user_input)
        curr_amount -= expense
        print(f"Expense added. Current balance: $ {curr_amount}")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

print(f"\nFinal balance: $ {curr_amount}")
```
